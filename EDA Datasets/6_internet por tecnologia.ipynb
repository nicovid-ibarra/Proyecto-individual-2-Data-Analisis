**ADSL**: *Número de conexiones a Internet por ADSL*.es una tecnología de transmisión de datos de banda ancha que utiliza líneas telefónicas estándar de cobre para proporcionar acceso a Internet de alta velocidad.La tecnología ADSL es "asimétrica" debido a que permite una mayor velocidad de descarga de datos que de carga. Esta tecnología se ha utilizado ampliamente en áreas donde las conexiones de banda ancha más rápidas, como la fibra óptica, no están disponibles.

**CABLEMODEM**: *Número de conexiones a Internet por cable módem*. El cable módem es un dispositivo que permite a los usuarios de servicios de televisión por cable tener acceso a Internet de alta velocidad. Los servicios de Internet a través de cable módem suelen ofrecer velocidades de conexión más rápidas que las tecnologías de conexión tradicionales como el DSL (Digital Subscriber Line) y el dial-up. Además, es una opción popular en áreas donde no está disponible la fibra óptica

**FIBRA OPTICA**: *Número de conexiones a Internet por fibra óptica*. La fibra óptica es una tecnología de transmisión de datos que utiliza hilos delgados y flexibles hechos de vidrio o plástico para transmitir datos en forma de pulsos de luz. En el contexto de las telecomunicaciones, la fibra óptica se utiliza para transmitir señales de Internet, televisión y telefonía a largas distancias con velocidades de transmisión extremadamente rápidas y con un alto ancho de banda.

**WIRELESS**: *Número de conexiones a Internet inalámbricas*. El término "wireless" en el contexto de las tecnologías de conexión a Internet generalmente se refiere a la conectividad inalámbrica que permite la transmisión de datos a través de ondas de radio en lugar de cables físicos. En este contexto, "wireless" puede referirse a una variedad de tecnologías inalámbricas, incluyendo Wi-Fi, Bluetooth y otras tecnologías de comunicación de corto y largo alcance.

Procedemos a multiplicar por 1000 las columnas correspondientes..

```python
# Multiplicar por 1000 las celdas en las columnas cuando sus valores sean menores a 12
columnas_a_multiplicar = ['ADSL', 'Cablemodem', 'Fibra óptica', 'Wireless', 'Otros', 'Total']

# Función para multiplicar por 1000 si el valor es menor a 12
def multiplicar_si_menor_a_12(valor):
    return valor * 1000 if valor < 12 else valor

# Aplicar la función a todas las celdas en las columnas especificadas
df2[columnas_a_multiplicar] = df2[columnas_a_multiplicar].applymap(multiplicar_si_menor_a_12)
```

Hacemos algunos graficos..

```python
# Crear una nueva columna 'Año-Trimestre' combinando la información de año y trimestre
df2['Año-Trimestre'] = df2['Año'].astype(str) + '-Q' + df2['Trimestre'].astype(str)

# Convertir la nueva columna 'Año-Trimestre' a objetos datetime
df2['Año-Trimestre'] = pd.to_datetime(df2['Año-Trimestre'], format='%Y-Q%m')

# Ordenar el DataFrame por la columna 'Año-Trimestre' en orden ascendente
df2 = df2.sort_values('Año-Trimestre')

# Seleccionar las columnas relevantes
columnas_evolucion = ['ADSL', 'Cablemodem', 'Fibra óptica', 'Wireless', 'Otros', 'Total']

# Graficar la evolución en el tiempo
plt.figure(figsize=(12, 8))

for columna in columnas_evolucion:
    plt.plot(df2['Año-Trimestre'], df2[columna], label=columna)

plt.title('Evolución en el Tiempo de Conexiones de Internet')
plt.xlabel('Año-Trimestre')
plt.ylabel('Cantidad')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

Se observa que los accesos en gral estan en alza excepto el ADSL por los motivos descritos en EDAS anteriores (se usa cada vez menos el acceso telefonico). Otro dato que se observa es un crecimiento constante de cablemodem y un crecimiento acelerado en el ultimo tiempo de fibra optica.

Se realiza graficos de torta para ver la distribucion de los tipos de tecnologia en 2014 y 2022 para ver contrastes

```python
# Seleccionar las columnas relevantes
columnas_torta = ['ADSL', 'Cablemodem', 'Fibra óptica', 'Wireless', 'Otros']

# Obtener los datos para el primer trimestre de 2014
datos_primer_trimestre_2014 = df2[(df2['Año'] == 2014) & (df2['Trimestre'] == 1)][columnas_torta].iloc[0]

# Obtener los datos para el cuarto trimestre de 2022
datos_cuarto_trimestre_2022 = df2[(df2['Año'] == 2022) & (df2['Trimestre'] == 4)][columnas_torta].iloc[0]

# Crear una figura con dos subgráficos
fig, axs = plt.subplots(1, 2, figsize=(15, 6))

# Primer subgráfico para el primer trimestre de 2014
axs[0].pie(datos_primer_trimestre_2014, labels=datos_primer_trimestre_2014.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
axs[0].set_title('1er Trimestre 2014')

# Segundo subgráfico para el cuarto trimestre de 2022
axs[1].pie(datos_cuarto_trimestre_2022, labels=datos_cuarto_trimestre_2022.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
axs[1].set_title('4to Trimestre 2022')

# Mostrar la figura
plt.show()

```

Se observa el claro crecimiento de cablemodem y fibra optica en detrimento del adsl


# EDA del archivo Accesos a Internet por tecnologia (No se usa en el analisis final)

```python
#importo 
ruta_al_archivo = '../DATA/06-Internet_Accesos por tecnologia.csv'
df = pd.read_csv(ruta_al_archivo, encoding='utf-8')
```

```python
#observo resultado
df.head()
```

```python
#Se ve la composicion
df.info()#hay 1 nulo
```

```python
df=df.drop_duplicates()
df=df.dropna()
```

```python
# Eliminar asteriscos de las columnas 'Año' y 'Trimestre'
df['Año'] = df['Año'].str.replace('*', '')
df['Trimestre'] = df['Trimestre'].str.replace('*','')

# Convertir las columnas 'Año' y 'Trimestre' a tipo numérico
df['Año'] = pd.to_numeric(df['Año'], errors='coerce')
df['Trimestre'] = pd.to_numeric(df['Trimestre'], errors='coerce')

# Crear una nueva columna 'Fecha' combinando 'Año' y 'Trimestre'
df['Fecha'] = pd.to_datetime(df['Año'].astype(str) + '-' + ((df['Trimestre'] - 1) * 3 + 1).astype(str), format='%Y-%m')
```

### Columna ADSL

```python
df_sorted = df.sort_values(['Año', 'Trimestre'])
# Crear el gráfico de evolución de ADSL para cada provincia con años ordenados
provincias = df['Provincia'].unique()

plt.figure(figsize=(12,6))

for provincia in provincias:
    datos_provincia = df_sorted[df_sorted['Provincia'] == provincia]
    plt.plot(datos_provincia['Año'].astype(str) + ' - Q' + datos_provincia['Trimestre'].astype(str), datos_provincia['ADSL'], marker='o', label=provincia)

plt.title('Evolución de ADSL en el tiempo por provincia')
plt.xlabel('Año - Trimestre')
plt.ylabel('ADSL')
plt.legend(loc='upper left', bbox_to_anchor=(1,1))
plt.grid()
plt.xticks(rotation=90)  # Cambia el ángulo de rotación a 90 grados para las etiquetas del eje x
plt.show()
```

Tendencia a la baja.
Gran cantidad en Capital,Cordoba y santa fe

```python
# Calcular el total de conexiones ADSL por provincia para corroborar colores del grafico anterior
total_adsl_por_provincia = df.groupby('Provincia')['ADSL'].sum()

# Crear un gráfico de barras para el total de ADSL por provincia
plt.figure(figsize=(10, 6))
total_adsl_por_provincia.plot(kind='bar', color='skyblue')
plt.title('Total de conexiones ADSL por provincia')
plt.xlabel('Provincia')
plt.ylabel('Total de conexiones ADSL')
plt.xticks(rotation=90)
plt.show()
```

### Columna cablemodem

```python
df_sorted = df.sort_values(['Año', 'Trimestre'])
# Crear el gráfico de evolución de Cablemodem para cada provincia con años ordenados
provincias = df['Provincia'].unique()

plt.figure(figsize=(12,6))

for provincia in provincias:
    datos_provincia = df_sorted[df_sorted['Provincia'] == provincia]
    plt.plot(datos_provincia['Año'].astype(str) + ' - Q' + datos_provincia['Trimestre'].astype(str), datos_provincia['Cablemodem'], marker='o', label=provincia)

plt.title('Evolución de Cablemodem en el tiempo por provincia')
plt.xlabel('Año - Trimestre')
plt.ylabel('Cablemodem')
plt.legend(loc='upper left', bbox_to_anchor=(1,1))
plt.grid()
plt.xticks(rotation=90)  # Cambia el ángulo de rotación a 90 grados para las etiquetas del eje x
plt.show()
```

Tendencia al alza.

```python
# Calcular el total de conexiones Cablemodem por provincia para corroborar colores del grafico anterior
total_adsl_por_provincia = df.groupby('Provincia')['Cablemodem'].sum()

# Crear un gráfico de barras para el total de ADSL por provincia
plt.figure(figsize=(10, 6))
total_adsl_por_provincia.plot(kind='bar', color='skyblue')
plt.title('Total de conexiones Cablemodem por provincia')
plt.xlabel('Provincia')
plt.ylabel('Total de conexiones Cablemodem')
plt.xticks(rotation=90)
plt.show()
```
