Esta tecnolog칤a se ha utilizado ampliamente en 치reas donde las conexiones de banda ancha m치s r치pidas, como la fibra 칩ptica, no est치n disponibles.

**CABLEMODEM**: *N칰mero de conexiones a Internet por cable m칩dem*. El cable m칩dem es un dispositivo que permite a los usuarios de servicios de televisi칩n por cable tener acceso a Internet de alta velocidad. Los servicios de Internet a trav칠s de cable m칩dem suelen ofrecer velocidades de conexi칩n m치s r치pidas que las tecnolog칤as de conexi칩n tradicionales como el DSL (Digital Subscriber Line) y el dial-up. Adem치s, es una opci칩n popular en 치reas donde no est치 disponible la fibra 칩ptica

**FIBRA OPTICA**: *N칰mero de conexiones a Internet por fibra 칩ptica*. La fibra 칩ptica es una tecnolog칤a de transmisi칩n de datos que utiliza hilos delgados y flexibles hechos de vidrio o pl치stico para transmitir datos en forma de pulsos de luz. En el contexto de las telecomunicaciones, la fibra 칩ptica se utiliza para transmitir se침ales de Internet, televisi칩n y telefon칤a a largas distancias con velocidades de transmisi칩n extremadamente r치pidas y con un alto ancho de banda.

**WIRELESS**: *N칰mero de conexiones a Internet inal치mbricas*. El t칠rmino "wireless" en el contexto de las tecnolog칤as de conexi칩n a Internet generalmente se refiere a la conectividad inal치mbrica que permite la transmisi칩n de datos a trav칠s de ondas de radio en lugar de cables f칤sicos. En este contexto, "wireless" puede referirse a una variedad de tecnolog칤as inal치mbricas, incluyendo Wi-Fi, Bluetooth y otras tecnolog칤as de comunicaci칩n de corto y largo alcance.

Procedemos a multiplicar por 1000 las columnas correspondientes..

```python
# Multiplicar por 1000 las celdas en las columnas cuando sus valores sean menores a 12
columnas_a_multiplicar = ['ADSL', 'Cablemodem', 'Fibra 칩ptica', 'Wireless', 'Otros', 'Total']

# Funci칩n para multiplicar por 1000 si el valor es menor a 12
def multiplicar_si_menor_a_12(valor):
    return valor * 1000 if valor < 12 else valor

# Aplicar la funci칩n a todas las celdas en las columnas especificadas
df2[columnas_a_multiplicar] = df2[columnas_a_multiplicar].applymap(multiplicar_si_menor_a_12)
```

Hacemos algunos graficos..

```python
# Crear una nueva columna 'A침o-Trimestre' combinando la informaci칩n de a침o y trimestre
df2['A침o-Trimestre'] = df2['A침o'].astype(str) + '-Q' + df2['Trimestre'].astype(str)

# Convertir la nueva columna 'A침o-Trimestre' a objetos datetime
df2['A침o-Trimestre'] = pd.to_datetime(df2['A침o-Trimestre'], format='%Y-Q%m')

# Ordenar el DataFrame por la columna 'A침o-Trimestre' en orden ascendente
df2 = df2.sort_values('A침o-Trimestre')

# Seleccionar las columnas relevantes
columnas_evolucion = ['ADSL', 'Cablemodem', 'Fibra 칩ptica', 'Wireless', 'Otros', 'Total']

# Graficar la evoluci칩n en el tiempo
plt.figure(figsize=(12, 8))

for columna in columnas_evolucion:
    plt.plot(df2['A침o-Trimestre'], df2[columna], label=columna)

plt.title('Evoluci칩n en el Tiempo de Conexiones de Internet')
plt.xlabel('A침o-Trimestre')
plt.ylabel('Cantidad')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

Se observa que los accesos en gral estan en alza excepto el ADSL por los motivos descritos en EDAS anteriores (se usa cada vez menos el acceso telefonico). Otro dato que se observa es un crecimiento constante de cablemodem y un crecimiento acelerado en el ultimo tiempo de fibra optica.

```python
# Seleccionar las columnas relevantes
columnas_torta = ['ADSL', 'Cablemodem', 'Fibra 칩ptica', 'Wireless', 'Otros']

# Obtener los datos para el primer trimestre de 2014
datos_primer_trimestre_2014 = df2[(df2['A침o'] == 2014) & (df2['Trimestre'] == 1)][columnas_torta].iloc[0]

# Obtener los datos para el cuarto trimestre de 2022
datos_cuarto_trimestre_2022 = df2[(df2['A침o'] == 2022) & (df2['Trimestre'] == 4)][columnas_torta].iloc[0]

# Crear una figura con dos subgr치ficos
fig, axs = plt.subplots(1, 2, figsize=(15, 6))

# Primer subgr치fico para el primer trimestre de 2014
axs[0].pie(datos_primer_trimestre_2014, labels=datos_primer_trimestre_2014.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
axs[0].set_title('1er Trimestre 2014')

# Segundo subgr치fico para el cuarto trimestre de 2022
axs[1].pie(datos_cuarto_trimestre_2022, labels=datos_cuarto_trimestre_2022.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
axs[1].set_title('4to Trimestre 2022')

# Mostrar la figura
plt.show()

```

# EDA del archivo Accesos a Internet por tecnologia (No se usa en el analisis final)

```python
#importo 
ruta_al_archivo = '../DATA/06-Internet_Accesos por tecnologia.csv'
df = pd.read_csv(ruta_al_archivo, encoding='utf-8')
```

```python
#observo resultado
df.head()
```

```python
#Se ve la composicion
df.info()#hay 1 nulo
```

```python
df=df.drop_duplicates()
df=df.dropna()
```

```python
# Eliminar asteriscos de las columnas 'A침o' y 'Trimestre'
df['A침o'] = df['A침o'].str.replace('*', '')
df['Trimestre'] = df['Trimestre'].str.replace('*','')

# Convertir las columnas 'A침o' y 'Trimestre' a tipo num칠rico
df['A침o'] = pd.to_numeric(df['A침o'], errors='coerce')
df['Trimestre'] = pd.to_numeric(df['Trimestre'], errors='coerce')

# Crear una nueva columna 'Fecha' combinando 'A침o' y 'Trimestre'
df['Fecha'] = pd.to_datetime(df['A침o'].astype(str) + '-' + ((df['Trimestre'] - 1) * 3 + 1).astype(str), format='%Y-%m')
```

### Columna ADSL

```python
df_sorted = df.sort_values(['A침o', 'Trimestre'])
# Crear el gr치fico de evoluci칩n de ADSL para cada provincia con a침os ordenados
provincias = df['Provincia'].unique()

plt.figure(figsize=(12,6))

for provincia in provincias:
    datos_provincia = df_sorted[df_sorted['Provincia'] == provincia]
    plt.plot(datos_provincia['A침o'].astype(str) + ' - Q' + datos_provincia['Trimestre'].astype(str), datos_provincia['ADSL'], marker='o', label=provincia)

plt.title('Evoluci칩n de ADSL en el tiempo por provincia')
plt.xlabel('A침o - Trimestre')
plt.ylabel('ADSL')
plt.legend(loc='upper left', bbox_to_anchor=(1,1))
plt.grid()
plt.xticks(rotation=90)  # Cambia el 치ngulo de rotaci칩n a 90 grados para las etiquetas del eje x
plt.show()
```

Tendencia a la baja.
Gran cantidad en Capital,Cordoba y santa fe

```python
# Calcular el total de conexiones ADSL por provincia para corroborar colores del grafico anterior
total_adsl_por_provincia = df.groupby('Provincia')['ADSL'].sum()

# Crear un gr치fico de barras para el total de ADSL por provincia
plt.figure(figsize=(10, 6))
total_adsl_por_provincia.plot(kind='bar', color='skyblue')
plt.title('Total de conexiones ADSL por provincia')
plt.xlabel('Provincia')
plt.ylabel('Total de conexiones ADSL')
plt.xticks(rotation=90)
plt.show()
```

### Columna cablemodem

```python
df_sorted = df.sort_values(['A침o', 'Trimestre'])
# Crear el gr치fico de evoluci칩n de Cablemodem para cada provincia con a침os ordenados
provincias = df['Provincia'].unique()

plt.figure(figsize=(12,6))

for provincia in provincias:
    datos_provincia = df_sorted[df_sorted['Provincia'] == provincia]
    plt.plot(datos_provincia['A침o'].astype(str) + ' - Q' + datos_provincia['Trimestre'].astype(str), datos_provincia['Cablemodem'], marker='o', label=provincia)

plt.title('Evoluci칩n de Cablemodem en el tiempo por provincia')
plt.xlabel('A침o - Trimestre')
plt.ylabel('Cablemodem')
plt.legend(loc='upper left', bbox_to_anchor=(1,1))
plt.grid()
plt.xticks(rotation=90)  # Cambia el 치ngulo de rotaci칩n a 90 grados para las etiquetas del eje x
plt.show()
```

Tendencia al alza.

```python
# Calcular el total de conexiones Cablemodem por provincia para corroborar colores del grafico anterior
total_adsl_por_provincia = df.groupby('Provincia')['Cablemodem'].sum()

# Crear un gr치fico de barras para el total de ADSL por provincia
plt.figure(figsize=(10, 6))
total_adsl_por_provincia.plot(kind='bar', color='skyblue')
plt.title('Total de conexiones Cablemodem por provincia')
plt.xlabel('Provincia')
plt.ylabel('Total de conexiones Cablemodem')
plt.xticks(rotation=90)
plt.show()
```
